### Vektorfält

Att rita vektorfält görs i huvudsak med kommandot `quiver()` som betyder pilkorg eller koger.
Kommandot har formen `quiver(x,y,u,v)`. Detta är dock ej den fullständiga formen med alla valmöjligheter.
x och y anger läget för pilens fot (detta kan också ställas in).
u och v anger riktningen x=1 y=1 och x=2 y=2 är samma riktning. Vektorns storlek skalas automatiskt av funktionen. Naturligtvis kan man komma åt skalningen av vektorerna om man vill.

```python
import numpy as np
import matplotlib.pyplot as plt

#pilens värden
x_position = 0
y_position = 0
x_direction = 4
y_direction = 4

plt.quiver(x_position, y_position, x_direction, y_direction)

plt.show()
```


Nedan ritas vektorfältet $ \overline{F}=\left(1,-1\right) $.
Vi använder kommandet `meshgrid()`. Det fungerar genom att ange intervall för x respektive y som sedan kombineras ihop till ett rutnät med punkter precis som i en Cartesisk produkt. A och B är mängder.

$$ A\times B=\left\{ \left(a,b\right):a\in A\land b\in B\right\} $$

I vårt fall ges A av `linspace(-5,5,10)`. A börjar med -5 och slutar med 5 och innehåller 10 tal (-5, -3,88..., -2,77...). Steglängden är 10/9=1,111...


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-5,5,10),np.linspace(-5,5,10))

#Vectorfield
u = 1
v = -1

plt.quiver(x,y,u,v)
plt.show()
```

Ett fält med mycket flödeut ur cirkeln:

$ \overline{F}=\left(x,y\right) $

Vi beräknar kurvintegralen (skalärt med tangenten) över enhetscirkeln och använder Greens sats

$
\int\overline{F}\bullet d\overline{r}	=\int Pdx+Qdy=\iint\left(\frac{\partial Q}{\partial x}-\frac{\partial P}{\partial y}\right)dxdy=\iint\frac{\partial(y)}{\partial x}-\frac{\partial(x)}{\partial y}\,dxdy
	=\iint0-0\,dxdy=0.
    $

Kurvintegralen är 0 och det kan förstås ur att flödet är hela tiden vinkelrät mot tangenten.

Om vi istället beräkna flödet (skalärt med normalen) så är det skilt från 0:

$
\int\overline{F}\bullet\overline{N}ds	=\iint\left(\frac{\partial P}{\partial x}+\frac{\partial Q}{\partial y}\right)dxdy=\iint\frac{\partial(x)}{\partial x}+\frac{\partial(y)}{\partial y}dxdy=\iint1+1\,dxdy
	=2\iint dxdy=2\pi.
    $

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-2,2,9),np.linspace(-2,2,9))

#Vektorfält
u = x
v = y

#Cirkel
t=np.linspace(0,2*np.pi,100)
xc=np.cos(t)
yc=np.sin(t)

plt.axis('equal')
plt.quiver(x,y,u,v)
plt.plot(xc,yc)
plt.show()
```

Vi studerar ett fält med mycket rotation men inget flöde.

$ \overline{F}=(y,-x) $

$
\int\overline{F}\bullet d\overline{r}	=\int Pdx+Qdy=\iint\left(\frac{\partial Q}{\partial x}-\frac{\partial P}{\partial y}\right)dxdy=\iint\left(\frac{\partial(-x)}{\partial x}-\frac{\partial(y)}{\partial y}\right)dxdy=
	\iint\left(-1-1\right)dxdy=-2\iint dxdy=-2\pi.$
    
$
\int\overline{F}\bullet\overline{N}ds	=\iint\left(\frac{\partial P}{\partial x}+\frac{\partial Q}{\partial y}\right)dxdy=\iint\left(\frac{\partial(y)}{\partial x}+\frac{\partial(-x)}{\partial y}\right)dxdy=
	\iint\left(0+0\right)dxdy=0.$

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

x,y = np.meshgrid(np.linspace(-2,2,9),np.linspace(-2,2,9))

#Vektorfält
u = y
v = -x

#Cirkel
t=np.linspace(0,2*np.pi,100)
xc=np.cos(t)
yc=np.sin(t)

plt.axis('equal')
plt.quiver(x,y,u,v)
plt.plot(xc,yc)
plt.show()
```
