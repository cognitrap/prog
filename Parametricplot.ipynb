# Parameterframställning av kurvor i 2D och 3D

## Enkel form av parametrisering

Inget speciellt behöver göras mer än att generera typiskt värden för en parameter t med linspace.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
#Parameter
t=np.linspace(0, 5, 50)

#(x(t),y(t))
x=t
y=60*np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
#Parameter
t=np.linspace(0, 2*np.pi, 100)

#(x(t),y(t))
x=np.sin(2*t+1.7)
y=np.sin(3*t)

plt.plot(x,y)
plt.show()
```

```python
#Parameterframställning av en astroid
t=np.linspace(0, 2*np.pi, 100)

#(x(t),y(t))
x=(np.sin(t))**3
y=(np.cos(t))**3

plt.plot(x,y)
plt.show()
```

```python
#Parameterframställning
t=np.linspace(-1.3, 1.3, 100)

#(x(t),y(t))
x=t*(t*t-1)
y=t*t-1

plt.plot(x,y)

plt.show()
```

## Vi ritar nu med hjälp av ett symbolpaket i python, sympy

Den har lite andra inställningar.

```python
# 2D Exempel 1

from sympy import *    
from sympy.plotting import plot_parametric
from sympy.plotting import plot3d_parametric_line
import math

init_printing()        # for nice math output 

# uncomment for plot in separate window
#%matplotlib qt        

# uncomment for plot inline
%matplotlib inline  

#Anger att t är en variabel
t = symbols('t')

# Cirkel
x = cos(t)
y = sin(t)
plot_parametric(x, y, (t, 0, 2*pi))
```

```python
# 2D Exempel 2
x = t + 0.5 * cos(10*t)
y = t + 0.5 * sin(10*t)
plot_parametric(x, y, (t, -2*pi, 2*pi))
```

```python
# 2D Exempel 3  Cycloid
x = t - sin(t)
y = 1 - cos(t)
plot_parametric(x, y, (t, -2*pi, 2*pi))
```

```python
# 2D  Exempel 4 
x = 0.1*t*cos(t)
y = 0.1*t*sin(t)
plot_parametric(x, y, (t, 0, 10*pi))
```

```python
import numpy as np
import matplotlib.pyplot as plt

an = np.linspace(0, 2*np.pi, 100)

plt.subplot(121)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.title('Olika skalade axlar', fontsize=10)

plt.subplot(122)
plt.plot(3*np.cos(an), 3*np.sin(an))
plt.axis('equal')
plt.title('Lika skalade axlar', fontsize=10)
```

## Rotation av parametriserade kurvor i 2D

Parametriserade kurvor kan enkelt roteras med en rotationsmatris. Och med hjälp av en glidare kan vinkeln varieras.

Rotationsmatrisen ges av 
$$
R_{\theta}=\left(\begin{array}{rr}
\cos\theta & -\sin\theta\\
\sin\theta & \cos\theta
\end{array}\right).
$$

```python
from ipywidgets import interact
import numpy as np
import matplotlib.pyplot as plt


def f(R):
#(x(t),y(t))
    t=np.linspace(-5, 5, 50)
    x=t
    y=t**2
    plt.axis([-2,2,-10,10])
    #plt.axis('equal')
    xrot=x*np.cos(R)-y*np.sin(R)
    yrot=x*np.sin(R)+y*np.cos(R)
    plt.plot(x,y)
    plt.plot(xrot,yrot)
    plt.show()
    return(R)

interact(f, R=(0, 2*np.pi,0.02))
```

```python
from ipywidgets import interact
import numpy as np
from sympy import *    
from sympy.plotting import plot_parametric

def f(R):
#(x(t),y(t))
    t = symbols('t')
    x=t
    y=t**2
    xrot=x*np.cos(R)-y*np.sin(R)
    yrot=x*np.sin(R)+y*np.cos(R)
    plot_parametric((x, y, (t, -5, 5)), (xrot, yrot, (t, -5, 5)))
    return(R)


interact(f, R=(0, 2*np.pi,0.1))
```
