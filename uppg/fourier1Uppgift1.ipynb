```python
from sympy import summation, symbols, sin, pi
from sympy.plotting import plot


# I sympy måste vi noga ange det som ska vara symboler
i = symbols('i', integer=True)  # Symbol för heltal
x = symbols("x")  # Variabel i funktionen

# Här följer den summa av fouriertermer jag vill beräkna, se ovan. Satt summan från 1 till 20.
SF = 4/pi*(summation(1 / (2*i+1) * sin((2*i+1)*x), [i, 1, 20]))

print(SF)

plot(SF,(x,-1,1))
```
