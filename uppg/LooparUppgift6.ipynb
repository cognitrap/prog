Presenterar flera olika program. Det första är vad man ofta kallar 'brute force' dvs att man gör beräkningen enkel och rakt fram, det kan ta rätt mycket körtid men lite tänke-tid.

I det första programmet körs 2 loopar. Den ena från 1 till 100 (max) den andra från 1 till 100. Vi får då naturligtvis alla möjliga fall då $x+y+z \leq 100$. Men vi får också en massa fall som egentligen inte är intressanta att testa. Med testa menas satsen `z2/s == s`. För att fundera lite över antalet tester som görs så räknas de med räknaren `c`, och skrivs ut på slutet.

````python
# Programmet gör inte riktigt vad det ska göra men finns här ändå.
# Det testar många fler möjligheter och samma tre tal mer än en gång.

import math as m

max = 100
c = 0

for x in range(1, max + 1):
    for y in range(1, max + 1):
        z2 = x*x + y*y
        s = int(m.sqrt(z2))
        c += 1
        if z2/s == s:
            #print(x**2, y**2, z2)
            print(x, y, s)

print("Undersökt ", c, " tal")
````

Vi undersöker således $100 \cdot 100 = 10000$ tal.

I nästa program försöker vi göra det bättre. Om x=1 så behöver vi bara undersöka y=1 till y=99 så att summan av x och y är under 100. Om x=2 behöver vi bara undersöka y=1 till y=98 osv. Övre gränsen för y är max-x, vilket blir max-x+1 i range().

````python
# Lite bättre men x+y+z är ibland större än 100

import math as m

max = 100
c = 0
for x in range(1, max + 1):
    for y in range(1, max - x + 1):  # Ändrat denna gräns
        z2=x*x + y*y
        s=int(m.sqrt(z2))
        c += 1
        if z2/s == s:
            print(x, y, s)
print("Undersökt ", c, " tal")
````

Vi undersöker nu cirka hälften så många tal. Antalet ges av en aritmetisk talföljd. $99+98+97+...+52+51+50+49+48+...+3+2+1 = 100\cdot49+50=4950$


I nästa program lägger vi in en if-sats. Detta ger ett nytt test men är mindre krävande för datorn. Vi gör den i 2 varianter. En med 2 if och en med en och-sats.

````python
# Ännu lite bättre. Vi testar inte om det visar sig att z gör att vi får en summa större än 100
# Observera att vi även får med de som bara är multiplar av varandra tex. (4,3,5) och (8,6,10).

import math as m

max = 100
c = 0
l = 0

for x in range(1, max + 1):
    for y in range(1, max - x + 1):
        l += 1
        z2 = x*x + y*y
        z = int(m.sqrt(z2))
        if x+y+z <= max:
            c += 1
            if z2/z == z:
                print(x, y, z)
print("Undersökt ", c, " tal")
print("Loopat ", l, " tal")
````

Nu blev det bara 1513 tal om vi först sållar med ett enklare test. Det är fortfarande 4950 tal som på något sätt testas, men bara 1513 som gör det tidskrävande testet.

Man kan också tänka att x och y ska behandlas symmetriskt. Då behöver man bara undersöka x upp till 50 (max/2).

````python
import math as m

max = 100
c = 0
l = 0

for x in range(1, int(max/2)):
    for y in range(1, max - x + 1):
        l += 1
        z2 = x*x + y*y
        z = int(m.sqrt(z2))
        if x+y+z <= max:
            c += 1
            if z2/z == z:
                print(x, y, z)
print("Undersökt ", c, " tal")
print("Loopat ", l, " tal")
````

Man ser ingen ändring i antalet undersökta tal men man har ändå en mindre mängd i looparna.


Man kan också tänka att z kan inte vara mindre än x så då kan man räkna bort 2x i gränsen i stället för bara ett x. Antalet tester minskar inte men antalet loopade tal minskar.

````python
import math as m

max = 100
c = 0
l = 0

for x in range(1, int(max/2)):
    for y in range(1, max - 2*x + 1):
        l += 1
        z2 = x*x + y*y
        z=int(m.sqrt(z2))
        if x+y+z <= max:
            c += 1
            if z2/z == z:
                print(x, y, z)
print("Undersökt ", c, " tal")
print("Loopat ", l, " tal")
````
