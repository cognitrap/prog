```python
from numpy import *
from matplotlib.pyplot import *

axis([-4,4,-4,4])

#abs(x)
x=linspace(-4, 4, 100)
y=sin(x)
plot(x,y,'k', label='sin(x)')


# 1 term
y=x
plot(x,y,'r:', label='1 term')

# 2 termer
y=x-x**3/6
plot(x,y,'r.', label='2 termer')

# 3 termer
y=x-x**3/6+x**5/120
plot(x,y,'b', label='3 termer')


legend(loc='best', fontsize='medium')

show()
#savefig('plot.png')
```
