```python
import matplotlib.pyplot as plt
import numpy as np

#Värdetabell
x=np.linspace(-10, 10, 100)
y=np.abs(x)

plt.axis([-10,10,-1,10])

plt.plot(x,y)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Värdetabell
x=np.linspace(-10, 10, 100)
y=np.exp(-x)*np.sin(x)

plt.axis([-1,4,-1,1])

plt.plot(x,y)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Värdetabell
x=np.linspace(-20, 20, 200)
y=np.sin(x)/x

plt.axis([-15,15,-0.5,1.5])

plt.plot(x,y)


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Värdetabell
x=np.linspace(-2, 2, 2000)
y=np.sin(1/x)

plt.axis([-2,2,-1,1])
plt.plot(x,y)
plt.show()

x=np.linspace(-0.2, 0.2, 2000)
y=np.sin(1/x)
plt.axis([-0.1,0.1,-1,1])
plt.plot(x,y)
plt.show()
```
