```python
import matplotlib.pyplot as plt
import numpy as np

#Mätdata som skrivits in i listor
Lx=[1,3,11,24]
Ly=[190,1700,24600,115700]

#Justerar axlar
plt.axis([0,30,0,120000])

#Plottar datamängden
plt.plot(Lx, Ly, 'b+', label='Data') #blue, markör + tecken

#Enkel anpassning, finns fler möjligheter
#Har valt att bestämma två värden på A o sedan ta medelvärde
A1=Ly[-1]/Lx[-1]**2
A2=Ly[-3]/Lx[-3]**2
A=(A1+A2)/2

#Plottar grafen med 200 punkter
x=np.linspace(0, Lx[-1], 200)
y=A*x**2
plt.plot(x,y, label='y=Ax²')

plt.legend()

plt.show()
print("A= ",A)
```
