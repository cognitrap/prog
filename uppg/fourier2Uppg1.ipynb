```python
from scipy.fft import fft, fftfreq, fftshift

import matplotlib.pyplot as plt
import numpy as np

# Antal samplingspunkter, sample rate: datapunkter per sekund
n = 500
# Signalen varar mellan -d och d
d = 5

# Gauss
def gen_wave2(A):
    x=np.linspace(-d,d,n*(2*d), endpoint=False)
    y = 1/np.sqrt(4*np.pi*A)*np.exp(-x**2/4/A)
    return(x,y)

# Genererar en gaussisk funktion genom anrop
Adef = 0.0005
x,y = gen_wave2(Adef)

plt.subplots(2,2, figsize=(12,8))  # Lite större figurer

# Ritar funktionen
plt.subplot(2,2,1)
plt.scatter(x,y)
plt.title('A=0,005')


# Kommandot för att få ut fouriertransformen
yf = fftshift(fft(y)) # själva transformen
xf=fftshift(fftfreq(n*(2*d),1/n)) # skapar korrekt frekvensskala


# Vi plottar beloppet av koefficienterna, amplitud för olika frekvenser
plt.subplot(2,2,3)
plt.scatter(xf,np.abs(yf))


# Vi kör för ett annat värde på A

# Genererar en gaussisk funktion genom anrop
Adef=0.5
x,y=gen_wave2(Adef)

# Ritar funktionen
plt.subplot(2,2,2)
plt.scatter(x,y)
plt.title('A=0,5')


# Kommandot för att få ut fouriertransformen
yf = fft(y) #själva transformen
xf=fftfreq(n*(2*d),1/n) #skapar korrekt frekvensskala


# Vi plottar beloppet av koefficienterna, amplitud för olika frekvenser
plt.subplot(2,2,4)
plt.scatter(xf,np.abs(yf))

plt.show()
```
