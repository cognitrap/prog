```python
import matplotlib.pyplot as plt
import random as r
import numpy as np

#Vi låser till 10 element
x=np.linspace(-20,20,10)

#Vi måste arbeta med listor
x=list(x)
noise=[]
y=[]

#Vi delar upp det i två
for i in range(0,10):
    noise.append(r.random())
for i in range(0,10):
    y.append(2*x[i]+5+20*noise[i])
    
plt.plot(x,y,'o')


#Nu jämnar vi till kurvan
M=10*[0]
M[0]=y[0]
M[-1]=y[-1]
for i in range(1,9):
    M[i]=(y[i-1]+y[i]+y[i+1])/3.
plt.plot(x,M,'+-')
plt.show()
```

Denna teknik kan tex. utföras innan man gör en anpassning med polyfit(). Anpassningen görs då till den nya listan beräknad med glidande medelvärde.
