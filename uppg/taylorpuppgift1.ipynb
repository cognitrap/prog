Termerna i polynomet ges av $$ \frac{x^{2k-1}}{(2k-1)!} $$

där $k=1,2,3...$.

Summan av termer upp till $n$ konstrueras av en def som i princip innehåller en loop. Fakulteten görs inte rekursiv.

```python
import numpy as np
import matplotlib.pyplot as plt

#funktion för fakultet
def fak(s):
    f=1
    for v in range(2,s+1):
        f=f*v
    return(f)

#summa av m termer; polynom
def taylor_sum(x,m):
    sumterm=0
    for i in range(1,m+1):#1 till m
        sumterm=(-1)**(i+1)*x**(2*i-1)/fak(2*i-1)+sumterm
    return(sumterm)
        

#skapar x- och y-värden för att rita polynom
Lx=list(np.arange(-2,2,0.2))
Ly=[]
for xi in Lx:
    Ly.append(taylor_sum(xi,3))


#Värdetabell för funktionen
xr=np.linspace(-2*np.pi, 2*np.pi, 100)
yr=np.sin(xr)

#Nu ritar vi!
plt.plot(xr,yr)

plt.plot(Lx,Ly)
plt.show()
```
