```python
from sympy import Heaviside, nan, symbols, sin, cos
from sympy.abc import x
from sympy.plotting import plot

x = symbols("x")

def H(x):
    return Heaviside(x)

a = -3; b = 3
def FK(x):
    return Heaviside(x - a)- Heaviside(x - b)

def Tsin(x):
    return FK(x)*sin(x)


# Nya förslag finns på hur label och legend ska anropas, men vi tar de gamla (1 juli 2022)

plot(H(x), (x,-4,4), label="Heaviside", legend=True)
plot(FK(x), (x,-4,4), label="Fyrkant", legend=True)
plot(Tsin(x), (x,-4,4), label="Kurva", legend=True)
```
