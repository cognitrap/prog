Ett plan

````python
# importavdelningen
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
import numpy as np

#för framtida användning
#import matplotlib
#%matplotlib ipympl


# lite bekvämlighet
fig = plt.figure()
ax = plt.axes(projection='3d')

# parametrarna t och s
tl=np.linspace(-20, 20, 1000)
sl=np.linspace(-20, 20, 1000)

# plane
s,t=np.meshgrid(sl,tl) #Kartesiska produkten
x = 2+t-s
y = -3+t+2*s
z= 1+t+s

# plotta
ax.plot_surface(x,y,z,linewidth=0, antialiased=False, cmap="coolwarm")

# visa plotten
plt.show()
````

### Rita vektorer

För att rita vektorer behöver vi ett nät av punkter att lägga dem i. Punkterna för detta anges av x och y. Det är förinställt att vektorns fot ska ligga här.

Sedan behöver vi ange själva vektorn. Vektorns x-komponent heter u och dess y-komponent heter v.

````python
import numpy as np
import matplotlib.pyplot as plt

# Punkter där man kan lägga vektorn, en sorts meshgrid
x = np.linspace(0,1,11)
y = np.linspace(0,1,11)

# Lägger i princip noll-vektorer i alla punkter
u = np.zeros((11,11))
v = np.zeros((11,11))

# Men vi vill ha en vektor som inte är en noll-vektor
# I punkten med index (3,5) lägger vi en vektor (0.2, 0.4)

u[3,5] = 0.2
v[3,5] = 0.4

#print(u,v)

plt.axis('equal') #axlar lika stora

#scale=1 behövs för annars gör python automatisk skalning
plt.quiver(x, y, u, v, scale=1, units='xy')
plt.show()
````

Observera problemet med angivelsen av vektorn i listorna u och v.

```
u[3,5] = 0.2
v[3,5] = 0.4
```

u anger vektorns storlek i x-led. Men [3,5] är index för matrisen. Så 3 anger raden med index 3 och 5 anger kolonnen med index 5, se figur.





Om man vill ha flera vektorer anges de med listor. Första elementet i varje lista är för vektor 1, andra elementet för vektor 2 osv. I koden ritas vektorer med utgångslägena (0,0), (0.2,0.3) och (0.2,0.1). Riktningen anges på motsvarande sätt. Riktningen för vektorn i punkten (0,0) ges av (0.2,0.1)

````python
import numpy as np
import matplotlib.pyplot as plt

#för framtida bruk
#import matplotlib
#%matplotlib ipympl
fig, ax = plt.subplots()

#start
x0=[0., 0.2, 0.2]
y0=[0., 0.3, 0.1]
dx=[0.2, 0.2, 0.3]
dy=[0.1, 0.1, 0.1]

plt.axis('equal') #axlar lika stora
ax.axis([-0.2,0.6,-0.2,0.6])


#scale=något tex 1; behövs för annars gör python automatisk skalning
plt.quiver(x0, y0, dx, dy, scale=1, units='xy')
plt.show()
````
