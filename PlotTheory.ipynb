# Objekt orienterad plottning

För att förstå plottning enligt OO så bör man först se hur ytan som det ritas i är uppdelat hierarkiskt.


![Overview](./Pictures/fig_map.png)



Ytterst har vi Figure som är den yttersta behållaren. Innanför den har vi behållaren Axes (med e). Man kan ha flera Axes i en Figure. Innanför Axes finns Axis (med i). Och innerst finns Ticks (markörer på axlarna) som inte är markerade i Axis.

När man ska rita skapar man först en Figure, sedan en eller flera Axes i den. Därefter bestämmer man Axis och Ticks.

Figure är tex. ett fönster eller en widget.
Axes är den yta som adresseras genom x- och y-koordinater (eller andra koordinater, även 3D och polära).

Enklast är att skapa en Figure med Axes genom kommandot pyplot.subplots eller plt.subplots. Sedan använder vi Axes.plot för att rita data.



````python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()  # Skapar en Figure med namnet 'fig', och en Axes med namnet ax.
ax.plot([1, 2, 3, 4], [1, 4, 2, 3])  # Vi ritar i Axesen ax. Skala och annat sköts automatiskt.
````

Naturligtvis är det inget speciellt med de engelska valen av ord. 

````python
import matplotlib.pyplot as plt

# (fig, ax = )
rita, graf = plt.subplots()
graf.plot([1, 2, 3, 4], [1, 4, 2, 3])
````

Observera att `plt.subplots()` ger 2 värden, både en Figure och en Axes.
Faktum är att man kan direkt skriva `plt.plot([1, 2, 3, 4], [1, 4, 2, 3])` och en Axes skapas automatiskt. Men vi ska försöka ha kontroll.

Kanske har du inte arbetat med att skriva objektorienterat tidigare. Du kan då tänka på Figure som en viss typ av dataobjekt. Vi ser också att vi inte har några objekt av typen Axes. Vi har ett objekt av typen Figure med storleken i pixlar.

````python
import matplotlib.pyplot as plt

fig=plt.figure() #Ger oss något att rita på
ax=plt.axes() #Ger oss ett objekt Axes med förinställda värden på Axis
````

Alla objekt, även behållare, är av typen Artist. Vi skapar sakerna var för sig stegvis.

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)

fig1=plt.figure() #fig1 är en Figure
ax1=fig1.add_subplot(1,1,1) #ax1 är en Axes

# Vi gör en ny Figure med plats för 2 grafer
fig2=plt.figure() #fig2 är en ny Figure
ax2=fig2.add_subplot(2,1,1) #ax2 är en Axes i fig2
ax3=fig2.add_subplot(2,1,2) #ax3 är en annan Axes i fig2

# Vi plottar i de 3 olika Axes
ax1.plot(x, np.cos(x), label="cos(x)")
ax2.plot(x, np.sin(x), label="sin(x)")
ax3.plot(x, np.sin(x)*np.cos(x), label="sin(x)cos(x)") #Vi lägger 2 i denna
ax3.plot(x, np.sqrt(x), label="sqrt(x)")

#Teckenförklaring
ax1.legend()
ax2.legend()
ax3.legend()

# Namn på axlar
ax1.set_xlabel('Tid')  # Add an x-label to the axes.
ax1.set_ylabel('Elongation')  # Add a y-label to the axes.
ax1.set_title("Pendel")  # Add a title to the axes.

ax2.set_ylabel('Energi')  # Add a y-label to the axes.
ax2.set_title("Fjäder")  # Add a title to the axes.

ax3.set_xlabel('Läge')  # Add an x-label to the axes.
ax3.set_ylabel('Hastighet')  # Add a y-label to the axes.
````

Vi skapar sakerna fortfarande var för sig men inte med `.add_subplot`

````python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2*np.pi, 100)

fig1, ax1 = plt.subplots(1,1)

# Vi gör en ny Figure med plats för 2 grafer. Observera att vi får direkt båda Axes.
fig2, (ax2,ax3) = plt.subplots(2,1) 

# Vi plottar i de 3 olika Axes
ax1.plot(x, np.cos(x), label="cos(x)")
ax2.plot(x, np.sin(x), label="sin(x)")
ax3.plot(x, np.sin(x)*np.cos(x), label="sin(x)cos(x)") #Vi lägger 2 i denna
ax3.plot(x, np.sqrt(x), label="sqrt(x)")

#Teckenförklaring
ax2.legend()
ax3.legend()

# Namn på axlar
ax1.set_xlabel('Tid')  
ax1.set_ylabel('Elongation')  
ax1.set_title("Pendel") 

ax2.set_xlabel('Styrka')  
ax2.set_ylabel('Energi')  

ax3.set_xlabel('Läge')  
ax3.set_ylabel('Hastighet')  
````

Observera trängseln som uppkommer och att vi därför saknar en etikett (label) på x-axeln för ax2 i fig2.
