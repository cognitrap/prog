```python
import matplotlib.pyplot as plt
import numpy as np
import cmath as cm


r=float(input("Anger r"))
v=float(input("Ange v"))
# z=rexp(ivt)
#Värdetabell
t=np.linspace(-2*np.pi, 2*np.pi, 100)
a=r*np.cos(v*t)
b=r*np.sin(v*t)

yR=a
plt.subplot(2,2,1)
plt.plot(t,yR)
yI=b
plt.subplot(2,2,3)
plt.plot(t,yI)
phase=np.arctan(yI/yR)
plt.subplot(2,2,2)
plt.plot(t,phase)
ab=np.sqrt(a**2+b**2)
plt.subplot(2,2,4)
plt.plot(t,ab)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cmath as cm


r1=float(input("Anger r1"))
v1=float(input("Ange v1"))
r2=float(input("Anger r2"))
v2=float(input("Ange v2"))

# z=r1exp(v1it)+r2exp(v2it)
#Värdetabell
t=np.linspace(-2*np.pi, 2*np.pi, 100)
a=r1*np.cos(v1*t)+r2*np.cos(v2*t)
b=r1*np.sin(v1*t)+r2*np.sin(v2*t)

yR=a
plt.subplot(2,2,1)
plt.plot(t,yR)
plt.title("realdel")
yI=b
plt.subplot(2,2,3)
plt.plot(t,yI)
plt.title("imaginärdel")
phase=np.arctan(yI/yR)
plt.subplot(2,2,2)
plt.plot(t,phase)
plt.title("fas")
ab=np.sqrt(a**2+b**2)
plt.subplot(2,2,4)
plt.plot(t,ab)
plt.title("absolutbelopp")


plt.show()
```
