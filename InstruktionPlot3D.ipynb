#### Kurvor i 3D

Kurvor i 3D ritas genom att ange kurvan på parameterform. Endast en parameter krävs. Vi behöver en lista över $t$ för att kunna beräkna $x(t)$ och $y(t)$ för att kunna beräkna kurvan enligt receptet $(x(t),y(t),z(t))$. Ofta är $z(t)=t$.
Vi skapar t genom np.linspace(start, slut, antal värden). Vi importerar numpy som np.

````python
import numpy as np
t=np.linspace(1, 20, 100)
````

Ytor i 3D

Vi använder matplotlib för att plotta.
Vi behöver specifikt mplot3d: `from mpl_toolkits import mplot3d`
Till denna modul skickar man ett keyword-argument `projection='3d'` så erhålls axlar i 3 dimensioner.
Vi sätter samman dessa och erhåller ett 3 dimensionellt system, dock utan en graf.


````python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
fig = plt.figure()
ax = plt.axes(projection='3d')
````

fig=plt.figure() ger oss en ram. Man kan ändra form på denna men inte rita på den.


ax=plt.axes(projection='3d') är det vi ritar på, själva canvasen. 


Se 3.8 för fler detaljer.
En figure() kan ha flera axes() men en axes() kan inte ha flera figure(). En axes() innehåller i sin tur axis(), dvs själva koordinataxlarna. Obs. den ena stavas med e den andra med i.

Vi ritar ett plan $z=2x+3y+2$.

````python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm #färgskalesystem
fig = plt.figure() #handtag för ramen
ax = plt.axes(projection='3d') #handtag för canvasen

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = 2*X+3*Y+2
# Rita ytan. Obs hur vi använder ax för att lägga till kommandon för axlarna
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

plt.show()
plt.savefig("testplan.png")
````


Vi ritar en paraboloid.

````python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

fig = plt.figure(figsize=(10,10)) # Vi kan ange storleken på ramen
ax = plt.axes(projection='3d')
ax.axis([-6,6,-6,6])  # Vi anger axlarna xmin,xmax,ymin,ymax

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
Z = 2*X**2+3*Y**2+2
# Rita ytan. 
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)

plt.show()
plt.savefig("testplan.png")
````

````python

# Används ej! Anteckning enbart

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')
# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)
# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))
# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()

''''''''''
# importing libraries 
from mpl_toolkits import mplot3d 
import numpy as np 
import matplotlib.pyplot as plt 
  
# defining surface and axes 
x = np.outer(np.linspace(-2, 2, 10), np.ones(10)) 
y = x.copy().T 
z = np.cos(x ** 2 + y ** 3) 
  
fig = plt.figure() 
  
# syntax for 3-D plotting 
ax = plt.axes(projection ='3d') 
  
# syntax for plotting 
ax.plot_surface(x, y, z, cmap ='viridis', edgecolor ='green') 
ax.set_title('Surface plot geeks for geeks') 
plt.show() 
````
