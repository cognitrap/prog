```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import sympy as sp
from sympy.plotting import plot3d

def paraboloid(a,b,x0,y0): 
  x, y = sp.symbols('x y')
  z=((x**2)/a)+((y**2)/b)
  plan=z.subs([(x,x0),(y,y0)])+sp.diff(z,x).subs([(x,x0)])*(x-x0)+sp.diff(z,y).subs([(y,y0)])*(y-y0)
  plot3d(plan,z)

interactive(paraboloid, a=widgets.IntSlider(value=5, description='a', max=15, min=-15),b=widgets.IntSlider(value=5, description='b', max=15, min=-15),x0=widgets.IntSlider(value=5, description='x0', max=15, min=-15),y0=widgets.IntSlider(value=5, description='y0', max=15, min=-15))
```
