```python
#Till en tenta vt23

from sympy import *
#from sympy import summation, symbols, cos, sin, pi, sinh, exp
from sympy.plotting import plot


# I sympy måste vi noga ange det som ska vara symboler
i = symbols('i', integer=True)  # Symboler för heltal
x = symbols('x')  # Behöver lite fler symboler


# Här följer den summa av fouriertermer jag vill beräkna, se ovan. 
SF=sinh(pi)/pi*summation((-1)**i / (1-I*i)*exp(I*i*x), [i, -16, 16])

#print(SF)


#plot(SF,(x,-pi,pi))
mm=plot(SF,(x,-pi,pi))
mm.save("expon.png")
```
