```python
# Mall för enkel plottning

import matplotlib.pyplot as plt
import numpy as np

# Värdetabell
x = np.linspace(-2*np.pi, 2*np.pi, 100)
y1 = 0.2/(0.2**2 + x**2)
y2 = 0.8/(0.8**2 + x**2)
y3 = 2/(2**2 + x**2)

#Nu ritar vi!
plt.plot(x,y1, label="a=0.2")
plt.plot(x,y2, label="a=0.8")
plt.plot(x,y3, label="a=2")

plt.legend(loc='best', fontsize='medium')
plt.savefig('plotkernel.png')
plt.show()
```

```python
# Mall för enkel plottning

import matplotlib.pyplot as plt
import numpy as np

# Värdetabell
x = np.linspace(-2, 2, 100)
y1=np.cosh(x)
y2=np.sinh(x)
#Nu ritar vi!
plt.plot(x,y1, label="cosh(x)")
plt.plot(x,y2, label="sinh(x)")

plt.legend(loc='best', fontsize='medium')
plt.savefig('hyperbolic.png')
plt.show()
```
