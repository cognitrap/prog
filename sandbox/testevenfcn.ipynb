```python
from scipy.fft import fft, ifft, fftfreq, fftshift
import numpy as np
import matplotlib.pyplot as plt

# Datamängd i array
n=8
# udda funktion
x = np.linspace(-1,1,n, endpoint=False)
y = np.heaviside(x, 0.5)
print("Data", y)
plt.title("Data")
plt.scatter(x,y)
plt.show()

# Beräkna fouriertransformen
y1 = fft(y)
x1 = fftfreq(n,1/n)
print("\nTransf not shifted y1", y1)
print("\nTransf not shifted x1", x1)

yf = fftshift(y1)
xf = fftshift(x1)
print("\nTransformen", yf)
plt.title("FT scatter, imag")
plt.scatter(xf, yf.imag)
plt.show()

plt.title("FT scatter, real")
plt.scatter(xf, yf.real)
plt.show()

plt.title("lineplot, imag")
plt.plot(xf, yf.imag)
plt.show()


yinv = ifft(yf)
#print("\nInverstransformen", yinv)
#print("\nRealdelen av inverstransformen", yinv)
```

```python
###### KLADD

from scipy.fft import fft, fftfreq, fftshift
import matplotlib.pyplot as plt
import numpy as np

# Antal samplingspunkter, sample rate: datapunkter per sekund
n = 3000
# Signalens varaktighet är från -d till d
d = 3



def wave():
    x = np.linspace(-d,d,n*(2*d), endpoint=False)
    y = np.heaviside(x,0.5)
    return(x,y)

x,y = wave()

# Ritar funktionen som ska transformeras
plt.title("Ursprunget")
plt.axis([-4,4,-0.1,2])
plt.plot(x,y)
plt.show()


# Kommandot för att få ut fouriertransformen för en av funktionerna
yf = fftshift(fft(y)) # Själva transformen
xf = fftshift(fftfreq(n*(2*d),1/n)) # Skapar en vettig frekvensskala, behöver i regel justeras


plt.figure(figsize=(9, 4))
plt.title("Realdelen")
# Realdelen
plt.axis([-20,20,-1,10000])  
plt.scatter(xf,yf.real)
plt.show()

# Imaginärdelen
plt.figure(figsize=(9, 4)) 
plt.title("Imaginärdelen")
plt.axis([-10,10,-3000,3000])  
plt.scatter(xf,yf.imag)
plt.show()

# Absolutbeloppet
plt.figure(figsize=(9, 4))  
plt.title("Absolutbeloppet")
plt.axis([-10,10,-10,400])  
plt.scatter(xf,abs(yf))
plt.show()
         
# Fasen
plt.figure(figsize=(9, 4))  
plt.title("Fasen")
plt.axis([-5,5,-5,5])  
plt.scatter(xf,np.angle(yf))
plt.show()
```
