```python
import matplotlib.pyplot as plt
import numpy as np

T=[0, 0.003, 0.007, 0.014, 0.02, 0.04, 0.08, 1]
x = np.linspace(0, 1, 100)
for t in T:
    y = np.exp(-np.pi**2*t)*np.sin(np.pi*x)+2*np.exp(-9*np.pi**2*t)*np.sin(3*np.pi*x)
    plt.plot(x,y)

plt.savefig("plot3_5solution.png")
plt.show()
```
